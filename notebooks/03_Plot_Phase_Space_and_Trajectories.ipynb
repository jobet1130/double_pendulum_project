```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path  
import matplotlib as mpl

mpl.rcParams['font.family'] = 'DejaVu Sans'
mpl.rcParams['mathtext.fontset'] = 'dejavusans'
sns.set_style("whitegrid")
%matplotlib inline 

DATA_PATH = Path("../data/processed/double_pendulum_processed.csv")
PLOT_DIR = Path("../plots")
```

```python
if not DATA_PATH.exists():
    raise FileNotFoundError(f"‚ùå Processed data not found at: {DATA_PATH}")

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["theta1"], df["omega1"], lw=1.0, color="royalblue")
plt.title(r"Phase Space: $\theta_1$ vs. $\omega_1$", fontsize=14)
plt.xlabel(r"$\theta_1$ (rad)")
plt.ylabel(r"$\omega_1$ (rad/s)")
plt.grid(True)
plt.tight_layout()

phase_theta1 = PLOT_DIR / "phase_space_theta1.png"
plt.savefig(phase_theta1, dpi=300)
plt.show()

print(f"‚úÖ Saved: {phase_theta1}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["theta2"], df["omega2"], lw=1.0, color="darkorange")
plt.title(r"Phase Space: $\theta_2$ vs. $\omega_2$", fontsize=14)
plt.xlabel(r"$\theta_2$ (rad)")
plt.ylabel(r"$\omega_2$ (rad/s)")
plt.grid(True)
plt.tight_layout()

phase_theta2 = PLOT_DIR / "phase_space_theta2.png"
plt.savefig(phase_theta2, dpi=300)
plt.show()

print(f"‚úÖ Saved: {phase_theta2}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["x2"], df["y2"], lw=1, color="purple")
plt.title("Trajectory of Second Pendulum Mass", fontsize=14)
plt.xlabel(r"$x_2$ (m)")
plt.ylabel(r"$y_2$ (m)")
plt.axis("equal")
plt.grid(True)
plt.tight_layout()

trajectory_plot = PLOT_DIR / "trajectory_plot.png"
plt.savefig(trajectory_plot, dpi=300)
plt.show()

print(f"‚úÖ Saved: {trajectory_plot}")
```
