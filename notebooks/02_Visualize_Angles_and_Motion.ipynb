```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from pathlib import Path 

sns.set_style("whitegrid")
%matplotlib inline
```

```python
data_path = Path("../data/processed/double_pendulum_processed.csv")

if not data_path.exists():
    raise FileNotFoundError(f"‚ùå Processed data not found at: {data_path}")

df = pd.read_csv(data_path)
df.head()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["time"], df["theta1"], label=r"$\theta_1$", color="royalblue")
plt.plot(df["time"], df["theta2"], label=r"$\theta_2$", color="tomato")
plt.title("Angular Position vs. Time", fontsize=14)
plt.xlabel("Time (s)")
plt.ylabel("Angle (rad)")
plt.legend()
plt.tight_layout()

plot_dir = Path("../plots")
plot_dir.mkdir(parents=True, exist_ok=True)
theta_plot = plot_dir / "angle_vs_time.png"
plt.savefig(theta_plot)
plt.show()

print(f"‚úÖ Saved: {theta_plot}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df["time"], df["omega1"], label=r"$\omega_1$", color="green")
plt.plot(df["time"], df["omega2"], label=r"$\omega_2$", color="orange")
plt.title("Angular Velocity vs. Time", fontsize=14)
plt.xlabel("Time (s)")
plt.ylabel("Angular Velocity (rad/s)")
plt.legend()
plt.tight_layout()

omega_plot = plot_dir / "omega_vs_time.png"
plt.savefig(omega_plot)
plt.show()

print(f"‚úÖ Saved: {omega_plot}")
```
